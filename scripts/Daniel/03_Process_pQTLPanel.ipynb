# pQTL panel

```r
library(readr)
library(dplyr)
library(tidyr)
library(purrr)
```

```r
projfolder <- '~/'
```

```r
olinkmap <- read_tsv(
  paste0(projfolder, 'data/olink_protein_map_3k_v1.tsv'),
  show_col_types = FALSE
)
head(olinkmap)
```

```r
olinkfilemap <- read_tsv(paste0(projfolder, 'data/olinkfilemapUKB.tsv'), show_col_types = FALSE)
head(olinkfilemap)
```

```r
olinkfilemap <- olinkfilemap %>%
  mutate(OlinkID = gsub('.*_(OID[^_]+)_.*', '\\1', name))
head(olinkfilemap)
```

```r
olinkIds <- read_tsv(
  paste0(projfolder, 'data/ProtIDsBN.tsv'),
  show_col_types = FALSE
)
head(olinkIds)
```

```r
protquery_olinkUKB <- olinkIds %>%
  inner_join(olinkmap, by = 'HGNC.symbol') %>%
  inner_join(olinkfilemap, by = 'OlinkID')
protquery_olinkUKB
```

```r
write_tsv(protquery_olinkUKB, paste0(projfolder, 'data/protquery_olinkUKB.tsv')) 
```
